```python
import pandas as pd
import os 
import json
from collections import defaultdict
```

```python
folder="odms_male_json"
dfJ=[]
```

```python
for files in os.listdir(folder):
    if files.endswith(".json"):
        path=os.path.join(folder,files)
        with open(path,"r",encoding='utf-8') as f:
            data=json.load(f)
        if isinstance(data,list):
            dfT=pd.DataFrame(data)
        if isinstance(data,dict):
            dfT=pd.DataFrame([data])
        
        dfJ.append(dfT)
```

```python
df=pd.concat(dfJ,ignore_index=True)
df.head()
```

```python
city=df['info'].apply(lambda x: x.get('city', None))
date = df['info'].apply(lambda x: x.get('dates', None))
outcome = df['info'].apply(lambda x: x.get('outcome', None))
pom = df['info'].apply(lambda x: x.get('player_of_match', None))
team1 = df['info'].apply(lambda x: x.get('teams', [None, None])[0])
team2 = df['info'].apply(lambda x: x.get('teams', [None, None])[1])
playing11Team1 = df['info'].apply(lambda x: list(x.get('players', {}).values())[0] if 'players' in x else None)
playing11Team2 = df['info'].apply(lambda x: list(x.get('players', {}).values())[1] if 'players' in x and len(x['players']) > 1 else None)
stadium=df['info'].apply(lambda x: x.get('venue', None))
event=df['info'].apply(lambda x: list(x.get('event', {}).values()))
event
```

```python
df[['city','date','outcome','pom','team1','team2','playing11Team1','playing11Team2','stadium','event']] = pd.DataFrame(list(zip(city,date,outcome,pom,team1,team2,playing11Team1,playing11Team2,stadium,event)))
df.drop(['info','meta'],axis=1,inplace=True)
df
```

```python
df['innings'][0]
```

```python
batsman_stats = defaultdict(lambda: {"runs": 0, "fours": 0, "sixes": 0})
bowler_stats = defaultdict(lambda: {"wickets": 0, "extras": 0, "fielders": []})
```

```python
for innings in df["innings"]:
    for inning in innings:
        for over in inning["overs"]:
            for delivery in over["deliveries"]:
                batter = delivery["batter"]
                bowler = delivery["bowler"]
                runs = delivery["runs"]["batter"]

                # Update batsman stats
                batsman_stats[batter]["runs"] += runs
                if runs == 4:
                    batsman_stats[batter]["fours"] += 1
                elif runs == 6:
                    batsman_stats[batter]["sixes"] += 1

                # Update bowler extras
                extras = delivery.get("extras", {})
                bowler_stats[bowler]["extras"] += sum(extras.values())

                # Update bowler wickets and fielders
                if "wickets" in delivery:
                    for wicket in delivery["wickets"]:
                        bowler_stats[bowler]["wickets"] += 1
batsman_stats
```

```python
playerStats = pd.DataFrame({'Batsman':batsman_stats,'Bowler':bowler_stats})
playerStats
```
